```python
from tkinter import *
from tkinter import Menu, filedialog, scrolledtext, messagebox

window = Tk()


window.title('Untitled Notepad')
window.geometry('600x400')

text = scrolledtext.ScrolledText(window,width=50,height=20, undo=True)
text.pack()

#Functions
def new():
    a = messagebox.askyesno('Notepad', 'The file is not saved \nDo you want a new notepad?')
    if a == YES:
        window.title('Untitled Notepad')
        file = None
        text.delete(1.0,END)
        
def open1():
    b = messagebox.askyesno('Notepad', 'Not saved \nDo you want to open files?')
    if b == YES:
        filedialog.askopenfile()
        
def save():
    filedialog.asksaveasfile(mode = 'w')

def exit():
    c = messagebox.askyesno('Notepad', 'The file is not saved \nDo you want to exit?')
    if c == YES:
        window.destroy()
        
def undo():
    text.edit_undo()
        
def cut():
    text.event_generate('<<Cut>>')
    
def copy():
    text.event_generate('<<Copy>>')
    
def paste():
    text.event_generate('<<Paste>>')
    
def select_all():
    text.tag_add('sel', 1.0, END)
        
#Menu
menu = Menu(window)

file = Menu(menu, tearoff = 0)
file.add_command(label='New', command = new)
file.add_command(label='Open', command = open1)
file.add_command(label='Save', command = save)
file.add_command(label='Save As', command = save)
file.add_separator()
file.add_command(label='Exit', command = exit)

edit = Menu(menu, tearoff = 0)
edit.add_command(label='Undo', command = undo)
edit.add_separator()
edit.add_command(label='Cut', command = cut)
edit.add_command(label='Copy', command = copy)
edit.add_command(label='Paste', command = paste)
edit.add_separator()
edit.add_command(label='Select All', command = select_all)

formt = Menu(menu, tearoff = 0)
formt.add_command(label='Word Wrap')
formt.add_command(label='Font...')

zoom = Menu(menu, tearoff = 0)
zoom.add_command(label='Zoom In')
zoom.add_command(label='Zoom Out')
zoom.add_command(label='Restore Default Zoom')
view = Menu(menu, tearoff = 0)
view.add_cascade(label='Zoom', menu=zoom)
view.add_command(label='Status Bar')

hlp = Menu(menu, tearoff = 0)
hlp.add_command(label='View Help')
hlp.add_command(label='Send Feedback')
hlp.add_separator()
hlp.add_command(label='About Notepad')

menu.add_cascade(label='File', menu=file)

menu.add_cascade(label='Edit', menu=edit)

menu.add_cascade(label='Format', menu=formt)

menu.add_cascade(label='View', menu=view)

menu.add_cascade(label='Help', menu=hlp)

window.config(menu=menu)







window.mainloop()
```
