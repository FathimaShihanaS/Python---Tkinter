```python
from tkinter import *

window = Tk()


window.title('Simple Calculator')
window.geometry('260x400')

entry_box= Entry(window,width=25)
entry_box.place(x=0 , y=0)


#FUNCTIONS
def click(no):
    old_number = entry_box.get()
    entry_box.delete(0,END)
    entry_box.insert(0, old_number + no)
    
def clear():
    entry_box.delete(0,END)
    
def addition():
    global first_number
    first_number = entry_box.get()
    global operation
    operation = 'addition'
    entry_box.delete(0,END)

def subtraction():
    global first_number
    first_number = entry_box.get()
    global operation
    operation = 'subtraction'
    entry_box.delete(0,END)

def multiplication():
    global first_number
    first_number = entry_box.get()
    global operation
    operation = 'multiplication'
    entry_box.delete(0,END)

def division():
    global first_number
    first_number = entry_box.get()
    global operation
    operation = 'division'
    entry_box.delete(0,END)
    
def equal():
    second_number = entry_box.get()
    entry_box.delete(0,END)
    if operation == 'addition':
        entry_box.insert(0, float(first_number) + float(second_number))
    elif operation == 'subtraction':
        entry_box.insert(0, float(first_number) - float(second_number))
    elif operation == 'multiplication':
        entry_box.insert(0, float(first_number) * float(second_number))
    elif operation == 'division':
        entry_box.insert(0, float(first_number) / float(second_number))
    
    
#BUTTONS
button1 = Button(text = '=', fg='white', bg='lightgreen', width=5, command = equal)
button1.place(x=195, y=355)

button2 = Button(text='+', fg='darkgreen', width=5, command = addition)
button2.place(x=195 , y=310)

button3 = Button(text='-', fg='darkgreen', width=5, command = subtraction)
button3.place(x=195 , y=265)

button4 = Button(text='x', fg='darkgreen', width=5, command = multiplication)
button4.place(x=195 , y=220)

button5 = Button(text='/', fg='darkgreen', width=5, command = division)
button5.place(x=195 , y=175)

button6 = Button(text='.', width=5, command= lambda: click('.'))
button6.place(x=130 , y=355)

button7 = Button(text='3', width=5, command= lambda: click('3'))
button7.place(x=130 , y=310)

button8 = Button(text='6', width=5, command= lambda: click('6'))
button8.place(x=130 , y=265)

button9 = Button(text='9', width=5, command= lambda: click('9'))
button9.place(x=130 , y=220)

button11 = Button(text='0', width=5, command= lambda: click('0'))
button11.place(x=65 , y=355)

button12 = Button(text='2', width=5, command= lambda: click('2'))
button12.place(x=65 , y=310)

button13 = Button(text='5', width=5, command= lambda: click('5'))
button13.place(x=65 , y=265)

button14 = Button(text='8', width=5, command= lambda: click('8'))
button14.place(x=65 , y=220)

button17 = Button(text='1', width=5, command= lambda: click('1'))
button17.place(x=0 , y=310)

button18 = Button(text='4', width=5, command= lambda: click('4'))
button18.place(x=0 , y=265)

button19 = Button(text='7', width=5, command= lambda: click('7'))
button19.place(x=0 , y=220)

button20 = Button(text='c', width=5, fg='red', command= clear)
button20.place(x=0 , y=355)


window.mainloop()
```
